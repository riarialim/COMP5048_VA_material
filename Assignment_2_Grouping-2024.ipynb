```python
print("Welcome to witnessing the squad formation of Planet COMP5048, Platoon {}.".format(interest_tut))

total_no_student = len(tut_namelist)
ideal_team_pax = 4
total_no_team = total_no_student//ideal_team_pax
five_pax_team = total_no_student%ideal_team_pax
four_pax_team = total_no_team - five_pax_team

# make sure all unikeys are lowercased
vip_list = [x.lower() for x in vip_list]
pair_list = [[x.lower() for x in sublist] for sublist in pair_list]

# case when too many vips
if len(vip_list) > five_pax_team:
    total_no_team = total_no_student//ideal_team_pax - 1
    five_pax_team += ideal_team_pax 
    four_pax_team = total_no_team - five_pax_team


# print("\nSome mathematics...")
print('\nTotal number of soldiers registered under Platoon {} : {}'.format(interest_tut,total_no_student),
      '\nTotal number of squads : ',total_no_team,
      '\nNumber of squads with 4 soldiers : ',four_pax_team,
      '\nNumber of squads with 5 soldiers (if any) : ',five_pax_team)

flat_pair = [item for sublist in pair_list for item in sublist]
# Check vip is in the tutorial namelist/ pair_list
for vip in vip_list:
    if vip not in tut_namelist:
        print(f">>>>>!!!!!Hellooooooooo {vip} doesnot seem to be in the classlist of {interest_tut}")
    if vip in flat_pair:
        print(f">>>>>!!!!!Hellooooooooo {vip} is in the nominated pair list, hence, will be removed from vip_list")
        vip_list.remove(vip)
        if len(vip_list)>0:
            print(f"Current vip(s): {vip_list}")
        else:
            print(f"Woohooo no vip in the Platoon")



# Check students are from this tutorial
def find_not_exist(flat_pair, l):
    error = []
    index = []
    i = 0
    for f in flat_pair:
        i +=1
        if f not in l:
            error.append(f)
            index.append(i)
    
    if error:
        print(flat_pair)
        return ">>>>>!!!!!Oh uh.. Suspected invader(s) detected :{} at position(s) {}\
        \nNomination only works for soldiers from the same territory\
        \nMake sure you give the correct unikeys.\n".format(error, index)
    
        
    else:
        return "Planet is safe!"

print('\nSearching for invaders >>>')
print('Result:\n',find_not_exist(flat_pair,tut_namelist))    


# Check if duplicate in the list
def find_dup(flat_pair):
    dupes = []
    index = []
    i = 0
    for f in flat_pair:
        i +=1
        if flat_pair.count(f) > 1:
            if f not in dupes:
                dupes.append(f)
                index.append(i)

    if dupes:
        print(pair_list)
        return "\n>>>>>!!!!!Warning! Clone(s) detected : {} at position(s) {}\
        \nKindly remind that the nomination needs to be two-way and establish a pair.\n".format(dupes, index)
    else:
        return "Wonderful, all nominations are genuine!"

print('\nSearching for clones in nominated pairs >>>')
print('Result:\n',find_dup(flat_pair))
    
```

## Auomated random grouping (Run All)

```python
# Codes for Grouping

flat_list = [item for sublist in pair_list for item in sublist] + vip_list

single_list = list(set(tut_namelist) - set(flat_list))

pairs_num = len(pair_list)

print('\nThe final nominated list of pairs: \n',pair_list)

print('\nNumber of pairs nominated : {}\n'.format(pairs_num))

print('\nChecking.. Single-but-will-never-walk-alone soldier(s):\n', single_list)

print('\nNumber of individual soldiers:\n{}\n'.format(len(single_list)))


print('\nSquads formation in progress...'.format(len(single_list)))
grouping = [[]]* total_no_team

# First Randomly assign the nominated pairs to groups 

if len(pair_list) >= total_no_team:
    for i in range(total_no_team):
        choice = random.choice(pair_list)
        grouping[i] = choice
        pair_list.remove(choice)
    if len(pair_list)>0:
        random.shuffle(pair_list)
        i_list = random.sample(range(0, total_no_team), len(pair_list))
        j = 0
        for i in i_list:
            grouping[i].extend(pair_list[j])
            j+=1

elif (len(pair_list) > 0) and (len(pair_list) < total_no_team):
    random.shuffle(pair_list)
    i_list = random.sample(range(0, total_no_team), len(pair_list))
    j = 0
    for i in i_list:
        grouping[i] = pair_list[j]
        j+=1
        
if pairs_num > 0:
    print("\nFirst stage of squad formation be like:\n",grouping)


# Assign Randomly those who have no nomination

# Make sure all groups has at least one member assigned:



while True:
    
#     print(grouping)
    if len(single_list) == 0:
        break
    else:
        # Check groups with vacancy. Prioritise those less than 4 members
        pax_in_group = []
        group_not_full = []
        i = 0
        for group in grouping:
            pax_in_group.append(len(group))
            if len(group) < ideal_team_pax:
                group_not_full.append(i)
            i+=1
            
        if len(group_not_full) == 0:
            break
        else:
            random.shuffle(single_list)
            j = random.choice(group_not_full)
            group_not_full.remove(j)
            if len(grouping[j]) == 0:
                grouping[j] = [single_list[-1]]
            else:
                grouping[j].append(single_list[-1])

            single_list.pop()
        continue

# pax_in_group = []    
# group_not_full = []
# i = 0
# for group in grouping:
#     pax_in_group.append(len(group))
#     if len(group) < ideal_team_pax:
#         group_not_full.append(i)
#     i+=1

if len(single_list) > 0:
    random.shuffle(single_list)
    i_list = random.sample(range(0, total_no_team), len(single_list))
    j = 0
    for i in i_list:
        if len(grouping[i]) == 0:
                grouping[i] = [single_list[j]]
        else:
            grouping[i].append(single_list[j])

        j+=1


# Assign vip as 5th member
if len(vip_list) > 0:
    
    print("\nTroops formation still in progress.. we're almost there :\n",grouping)
    
    for group in grouping:
        if len(vip_list) > 0 and len(group) < ideal_team_pax:
            vip = random.choice(vip_list)
            group.append(vip)
            vip_list.remove(vip)

    for group in grouping:
        if len(vip_list) > 0 and len(group)== ideal_team_pax:
            vip = random.choice(vip_list)
            group.append(vip)
            vip_list.remove(vip)

        
print("\nTadaaaa -- \nThe ultimate troops are born!\n\nPlanet {} is ready to fight!\n{}".format(interest_tut,grouping))
```

```python
# Record the result to a dataframe
df_group = []
tnum = 0
for group in grouping:
    tnum += 1
    group_name = interest_tut +'-G'+"{:02d}".format(tnum)
    modified_df = pd.DataFrame({group_name: group}) 
    df_group.append(pd.Series(modified_df[group_name], name=group_name))

df_group = pd.concat(df_group, axis=1)
    
df_group = pd.melt(df_group).dropna().reset_index(drop=True)
df_group.rename(columns={'variable':'group_name','value':'login_id'}, inplace=True)
df_group  
   
```

```python
# Merge with names

df_final = df[df['class']==interest_tut].reset_index(drop=True)
df_result = df_final.join(df_group.set_index('login_id'), on='login_id')
str2 = '_group.csv'
csv_name = f"{interest_tut}{str2}"

df_result.sort_values(by=['group_name','name']).reset_index(drop=True)
```
